```python
usa_rows = df.loc[df['Country'] == 'usa']
usa_rows
```

```python
# Eliminar las filas que contienen el valor 'unknown' en la columna 'Fatal'
df_1 = usa_rows.loc[usa_rows['Fatal (Y/N)'] != 'unknown']
df_1
```

```python
# Limpiar y convertir los valores en la columna "Fatal" a numéricos (1: True, 0: False)
df_1['Fatal (Y/N)'] = df_1['Fatal (Y/N)'].replace({'True': 1, 'False': 0}).astype(int)
df_1
```

```python
# Agrupar el DataFrame por el área geográfica y calcular el ratio de mortalidad promedio
mortality_ratio = df_1.groupby('Area')['Fatal (Y/N)'].mean()

# Crear una lista de etiquetas de área geográfica
areas = list(mortality_ratio.index)

# Obtener el ratio de mortalidad para cada área
mortality_values = [mortality_ratio[area] for area in areas]

# Crear el gráfico de barras para comparar el ratio de mortalidad por área geográfica
plt.bar(areas, mortality_values, width=0.8)
plt.xlabel('Área Geográfica')
plt.ylabel('Ratio de Mortalidad')
plt.title('Comparación del Ratio de Mortalidad en USA por Área Geográfica')
plt.xticks(rotation=80)

# Ajustar los parámetros left y right de los ejes
plt.subplots_adjust(left=0.1, right=1.1)

# Mostrar el gráfico
plt.show()

# Guardar el gráfico en un archivo en tu directorio de trabajo
fig = plt.gcf()
fig.savefig('grafico.png')
```
